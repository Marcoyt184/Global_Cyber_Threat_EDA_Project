```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/final_kaggle_cyber_dataset.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
df.isnull().sum()
```

```python
top_countries = df['country_name'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top Countries by Malicious IPs")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='risk_level', data=df)
plt.title("Risk Level Distribution")
plt.show()
```

```python
sns.countplot(x='time_zone_attack', data=df)
plt.title("Attack Time Zones")
plt.show()
```

```python
sns.countplot(x='reported_weekday', data=df,
              order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title("Attacks by Weekday")
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

country_counts = df['country_name'].value_counts().reset_index()
country_counts.columns = ['country', 'count']

fig = px.choropleth(country_counts,
                    locations="country",
                    locationmode="country names",
                    color="count",
                    title="Global Cyber Attacks by Country",
                    color_continuous_scale="Reds")

fig.show()
```
